```python
import random

from matplotlib import pyplot as plt
x = range(2,26,2)
y = [15,13,14,5,17,20,25,26,26,24,22,18]
plt.plot(x,y)
plt.show()
```

```python
#增加功能：1.设置图片大小
        #2.保存图片
```

```python
fig = plt.figure(figsize=(20,8),dpi=88)
x = range(2,26,2)
y = [15,13,14,5,17,20,25,26,26,24,22,18]
plt.plot(x,y)
plt.savefig("../img/numpy_pandas_matplotlib/ma01_1.jpg")
plt.show()
```

```python
fig = plt.figure(figsize=(20,8),dpi=88)
x = range(2,26,2)
y = [15,13,14,5,17,20,25,26,26,24,22,18]
plt.plot(x,y)
plt.xticks(x)
plt.savefig("../img/numpy_pandas_matplotlib/ma01_2.jpg")
plt.show()
```

```python
import random
times = range(0,120)
temperature = [random.randint(20,35) for i in range(120)]
plt.figure(figsize=(30,8),dpi=88)
plt.plot(times,temperature)
time_label = ["10:{}".format(i) for i in range(60)]
time_label += ["11:{}".format(i) for i in range(60)]
plt.xticks(times,time_label,rotation=45)#rotataion旋转刻度标签
plt.show()
```
