```python
import cv2
#创建窗口
cv2.namedWindow('new',cv2.WINDOW_NORMAL)#WINDOW_NORMAL可以更改窗口大小 WINDOW_AUTOSIZE不允许修改窗口大小
#更改窗口大小
cv2.resizeWindow('new',1920,1080)
#展示名为new的窗口
cv2.imshow('new',0)
key = cv2.waitKey(0)
if key == ord('q'):#python中计算ascii的函数,0表示接受任意按键，如果给其他数字代表等待案件的时间，单位毫秒
    exit()
    cv2.destroyAllWindows()
```

```python
#封装函数
def cv_show(name,img):
    cv2.imshow(name,img)
    key = cv2.waitKey(0)
    if key& 0xFF ==ord('q'):
        cv2.destroyAllWindows()
```

```python
cv_show('dog',cqy)
```

```python
img = cv2.imread('../img/cqy_smile_dataset/xiao_si.jpg')
while True:
    cv2.imshow('img',img)
    key = cv2.waitKey(0)
    if key & 0xFF == ord('q'):
        break
    elif key & 0xFF == ord('s'):
        cv2.imwrite('../img/opencv/test01.png',img)
    else: print(key)
cv2.destroyAllWindows()
```
