```python
import cv2
import math
path = 'F:\\cqy_smile_dataset\\xiao_rong_zhu_jian_can_lan.jpg'
img = cv2.imread(path)
pointList = []

def mousePoints(event,x,y,flags,params):
    if event == cv2.EVENT_LBUTTONDOWN:
        size = len(pointList)
        #只要点列表中点数不到三个或者不为空就开始绘制线段 绘制线段的两个点 分别来自刚刚点击生成的（x,y)和点列表中的
        if size != 0 and size % 3 != 0:
            cv2.line(img,tuple(pointList[round((size -1) / 3) * 3]), (x,y),(0,0,255),2)
        cv2.circle(img,(x,y),5,(0,0,255),cv2.FILLED)
        pointList.append([x,y])
        # print(pointList)

def gradient(pt1,pt2):
    return (pt2[1] - pt1[1]) / (pt2[0] - pt1[0])

def getAngle(pointList):
    pt1,pt2,pt3 = pointList[-3:]
    m1 = gradient(pt1,pt2)
    m2 = gradient(pt1,pt3)
    angle_R = math.atan((m2 -m1) / (1 + m2 * m1))
    angle_D = round(math.degrees(angle_R))
    cv2.putText(img,str(angle_D),(pt1[0]-40,pt1[1]-20),cv2.FONT_HERSHEY_COMPLEX,1.5,(0,0,255),2)
    print(angle_D)

while True:
    if len(pointList) == 3:
        getAngle(pointList)
    cv2.imshow('img',img)
    cv2.setMouseCallback('img',mousePoints)
    if cv2.waitKey(1) & 0xFF == ord('f'):
        pointList = []
        img = cv2.imread(path)
    elif cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
```
